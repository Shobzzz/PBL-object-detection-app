```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  #モデル

results = model.train(
    data="/content/drive/MyDrive/PBL-object-detection-app/object_detection/dataset-white/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    workers=4,
    patience=20,
    lr0=0.01,
    augment=True,
    cache=False,
    name="buoy_detection_v1",
    exist_ok=True,
)

print(f"Best weights saved at: {results.save_dir}/weights/best.pt")

```
